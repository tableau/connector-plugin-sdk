```python
list_of_files = []

for root, dirs, files in os.walk('tdvt/exprtests'):
#     for file in files:
#         if file.startswith('setup.'):
#             print(os.path.join(root, file))
    for file in files:
        if not 'expected' in file:
            list_of_files.append(os.path.join(root, file))
        
list_of_files.sort()
list_of_files
```

```python
# a dict of dir: 
# dir: {
#     file: {testcase: [args, args], testcase: [args],}, 
#     file2: {testcase: [args]}, 
#     {subdir: [{file: testcase}], ...}
# }

files_and_tests = {}
```

```python
# next iterate through the files, using regex to find the test cases
data_to_csv = []

for filename in list_of_files:
    test_suite = filename.split('/')[2]
    test_subsuit = None
    if len(filename.split('/')) > 3:
        test_subsuite = '/'.join(filename.split('/')[3:-1])
    test_name = filename.split('/')[-1].replace('setup.', '').replace('.txt', '')
#     print(filename + '\n', test_suite, test_subsuite, test_name)
    data_to_csv.append([test_suite, test_subsuite, test_name])
    
data_to_csv
```

```python
import csv

header = ['test suite', 'test subsuite', 'test case']

with open('/Users/lpetschauer/Desktop/test_suite_data.csv', 'w', encoding='UTF-8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    for l in data_to_csv:
        writer.writerow(l)
```

```python
#  our regex is something like: ([A-Z_]+)\((\w+)\)
# two capture groups; one for the function name, one for the contents

logical_files = []

for root, dirs, files in os.walk('/Users/lpetschauer/Downloads/connector-plugin-sdk-master/tdvt/tdvt/logicaltests/generate/input/'):
#     for file in files:
#         if file.startswith('setup.'):
#             print(os.path.join(root, file))
    for file in files:
        if not 'expected' in file:
            logical_files.append(os.path.join(root, file))
        
logical_files.sort()
logical_files
```

```python
logical_data_to_csv = []

for filename in logical_files:
    test_suite = filename.split('/')[9]
    test_subsuit = None
    if len(filename.split('/')) > 9:
        test_subsuite = '/'.join(filename.split('/')[10:-1])
    test_name = filename.split('/')[-1].replace('setup.', '').replace('.xml', '')
#     print(filename + '\n', test_suite, test_subsuite, test_name)
    logical_data_to_csv.append([test_suite, test_subsuite, test_name])
    
logical_data_to_csv
```
